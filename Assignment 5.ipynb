```python
# Import Libraries
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

```

```python
# Store filepaths in variables
city_data="city_data.csv"
ride_data="ride_data.csv"
```

```python
# Open files as Dataframes using pandas library reader module
city_df=pd.read_csv(city_data)
city_df.head()
```

```python
# Open files as Dataframes using pandas library reader module
ride_df=pd.read_csv(ride_data)
ride_df.head()
```

```python

Data_Merge=pd.merge(city_df, ride_df, on="city", how="outer")
Transport_1=Data_Merge.rename(columns={"city":"City", "driver_count":"Drivers", "type":"Type","date":"Date","fare":"Fare", "ride_id":"Ride_ID"})
Transport_1.head()
```

```python
# Finding Total Number of Rides Per City
Total_number_rides_per_city=Data_Merge["city"].value_counts()
List_Towns=Total_number_rides_per_city.index.tolist()
List_Towns
                                
```

```python
# Finding Average Fare per City 
Data_Index_Change=Data_Merge.set_index("city")
Town_Fare=[]
for x in List_Towns:
    Fare=Data_Index_Change.loc[x,"fare"]
    Town_Fare.append(Fare)
Town_Fare
    

```
